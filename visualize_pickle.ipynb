```python
import pandas as pd
import numpy as np

import results_calc
import zipfile
```

**Регулируемые параметры**

```python
SENSOR_INDEX = 2 #0, 1, 2, 3

#Тип кросс-валидации
# TYPE_OF_TEST = "TimeSeries"
# TYPE_OF_TEST = "fixed"
TYPE_OF_TEST = "CV"

#папки с путями до pikcle таблиц с результатами
FOLDERS = [f'{x}/{SENSOR_INDEX}' for x  in [
    'baseline_results',
    'gan_results'
] ]

#требования к размерами и количеству фолдов 
FOLDS_NUMBER = {"CV": {'train_size_in_folds':3, 'n_folds':6}, "fixed": {'n_folds':6}, "TimeSeries": {'n_folds':6}}
```

**Расчет метрик и построение графиков для выбранных параметров. Формирование сводной таблицы по разным моделям**

```python
results = results_calc.recalculate(folder=FOLDERS, plots_folder="results/plots", \
                type_of_test=TYPE_OF_TEST, FOLDS_NUMBER=FOLDS_NUMBER, verbose=True)
```

```python
#Лучшие по каждой модели
top_types = results.sort_values('rmse').groupby(['name']).head(1).sort_values('rmse')
top_types
```

**Вычисление лучших для каждой переменной для конкретного вида кросс-валидации. Формирование сводной таблицы и сохранения графиков**

```python
# TYPE_OF_TEST = "TimeSeries"
TYPE_OF_TEST = "fixed"
# TYPE_OF_TEST = "CV"

sensors = [0, 1, 2, 3]

best = []
for i in sensors:
    FOLDERS = [f'{x}/{i}' for x  in [
        'baseline_results',
        'gan_results'
    ] ]
    results = results_calc.recalculate(folder=FOLDERS, plots_folder="results/plots", \
                    type_of_test=TYPE_OF_TEST, FOLDS_NUMBER=FOLDS_NUMBER)
    res = results.sort_values('rmse').groupby(['name']).head(1)
    best.append(res)
    res['sensor'] = [i] * res.shape[0]
best = pd.concat(best, axis=0, ignore_index=True)
best = best[['sensor'] + list(best.columns[:-1])]
best = best.sort_values(['sensor', 'rmse'])
```

```python
best
```

```python
best.to_csv(f"results/best_{TYPE_OF_TEST}.csv", sep=";")
```

```python
paths = list(best['plot_path'])
zip_file = zipfile.ZipFile(f"results/best_plots_{TYPE_OF_TEST}.zip", "w")
for path in paths:
    zip_file.write(path)
zip_file.close()
```

```python
paths = list(best['table_path'])
zip_file = zipfile.ZipFile(f"results/best_tables_{TYPE_OF_TEST}.zip", "w")
for path in paths:
    zip_file.write(path)
zip_file.close()
```
